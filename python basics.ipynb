```python

# TOPIC 2: VARIABLES


# THEORY:
# A variable is used to store data.
# We assign values using the = operator.
# Variable names:
# - Cannot start with a number
# - Cannot contain spaces
# - Are case-sensitive

name = "Pragathi"
age = 21
job = "Engineer"
college = "BMSCE"

print(name)
print(age)
print(job)
print(college)
```

```python

# TOPIC 3: DATA TYPES


# THEORY:
# Python has different data types:
# 1. str  -> String (text)
# 2. int  -> Integer (whole number)
# 3. float -> Decimal number
# 4. bool -> True or False

name = "Pragathi"     # string
age = 21              # integer
height = 5.4          # float
is_engineer = True    # boolean

print(type(name))
print(type(age))
print(type(height))
print(type(is_engineer))
```

```python

# TOPIC 4: USER INPUT


# THEORY:
# input() function takes input from the user.
# It always returns data as a string.
# Use type casting (int()) to convert to integer.

name = input("Enter your name: ")
age = int(input("Enter your age: "))

print("Hello", name)
print("You are", age, "years old")
```

```python

# TOPIC 5: OPERATORS


# THEORY:
# Operators are used to perform operations on variables.
# Types:
# 1. Arithmetic (+, -, *, /, %, **)
# 2. Comparison (==, !=, >, <, >=, <=)
# 3. Logical (and, or, not)

a = 10
b = 5

# Arithmetic
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)

# Comparison
print("Is a greater than b?", a > b)

# Logical
age = 21
print("Is age between 18 and 30?", age > 18 and age < 30)
```

```python

# TOPIC 6: TYPE CASTING


# THEORY:
# Type casting means converting one data type into another.
# int() -> convert to integer
# float() -> convert to float
# str() -> convert to string

age_string = "21"

age_number = int(age_string)

print(age_number + 5)
print(type(age_number))
```

```python

# TOPIC 7: CONDITIONAL STATEMENTS (IF-ELSE)


# THEORY:
# Conditional statements allow decision making.
# Syntax:
# if condition:
#     block of code
# elif condition:
#     block of code
# else:
#     block of code

age = 21

if age >= 18:
    print("Pragathi is eligible to work as an Engineer.")
else:
    print("Not eligible to work.")

marks = 85

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
else:
    print("Grade C")
```

```python

# TOPIC 8: FOR LOOP


# THEORY:
# A loop is used to repeat code multiple times.
# for loop is used when we know the number of repetitions.
# range(n) generates numbers from 0 to n-1.

for i in range(5):
    print("Hello Pragathi - Iteration:", i)

for i in range(0, 6):
    print(i)

```

```python

# TOPIC 9: WHILE LOOP

# THEORY:
# while loop runs as long as the condition is True.
# It is used when the number of iterations is unknown.

count = 1

while count <= 5:
    print("Count is:", count)
    count += 1   # increment
```

```python

# TOPIC 10: LISTS

# THEORY:
# A list stores multiple values in a single variable.
# Lists are ordered and changeable.
# Defined using square brackets [].

skills = ["Python", "Engineering", "Problem Solving"]

print(skills[0])      # Access first element
skills.append("Git")  # Add new element

for elements in skills:
    print(elements)
```

```python

# TOPIC 11: TUPLES


# THEORY:
# A tuple is similar to a list but cannot be modified.
# Tuples are defined using parentheses ().

coordinates = (10, 20)

print(coordinates[0])
```

```python
# TOPIC 12: SETS


# THEORY:
# A set stores unique values.
# Duplicates are automatically removed.
# Defined using curly braces {}.

numbers = {1, 2, 3, 3, 4}

print(numbers)
```

```python

# TOPIC 13: DICTIONARIES


# THEORY:
# A dictionary stores data in key-value pairs.
# Defined using curly braces {}.
# Keys must be unique.

profile = {
    "name": "Pragathi",
    "age": 21,
    "job": "Engineer",
    "college": "BMSCE"
}

print(profile["name"])

for key, value in profile.items():
    print(key, ":", value)
```

```python

# TOPIC 14: STRINGS


# THEORY:
# Strings are sequences of characters.
# Strings are immutable (cannot be changed).
# Useful string methods:
# upper(), lower(), len()

name = "Pragathi"

print(name.upper())
print(name.lower())
print("Length:", len(name))
print("First letter:", name[0])
```
